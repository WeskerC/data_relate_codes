#### * LINES                                                                   *

#### See https://matplotlib.org/stable/api/artist_api.html#module-matplotlib.lines
#### for more information on line properties.
- lines.linewidth: 1.5               # line width in points
- lines.linestyle: -                 # solid line
- lines.color:     C0                # has no affect on plot(); see axes.prop_cycle
- lines.marker:          None        # the default marker
- lines.markerfacecolor: auto        # the default marker face color
- lines.markeredgecolor: auto        # the default marker edge color
- lines.markeredgewidth: 1.0         # the line width around the marker symbol
- lines.markersize:      6           # marker size, in points
- lines.dash_joinstyle:  round       # {miter, round, bevel}
- lines.dash_capstyle:   butt        # {butt, round, projecting}
- lines.solid_joinstyle: round       # {miter, round, bevel}
- lines.solid_capstyle:  projecting  # {butt, round, projecting}
- lines.antialiased: True            # render lines in antialiased (no jaggies)

```python
import matplotlib.pyplot as plt
```

```python
# 查看当前block内的绘图
plt.gcf()
```

```python
# plot，画图
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

#### plot的第三位参数用于设置线的种类

```python
#color, marker and linestyle.
#颜色，符号和线条类型
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'go--', linewidth=2, markersize=12)
#也等于  color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12)
#关键字比参数先行
plt.show()
```

##### 设置 x 和 y 轴的刻度范围

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'go--', linewidth=2, markersize=12)
# x 轴的最小值、最大值，y 轴的最小值、最大值
plt.axis([0, 6, 0, 20]) 
#如果数据点超出了这个范围，则会被截断并不会在图形中显示出来。
plt.show()
```

#### 通过数据结构的名称导入要绘制的数据
所有可以被编码的对象都支持则会中引入，例如字典，pandas数据结构或nunpy数组  
All indexable objects are supported. This could e.g. be a dict, a pandas.DataFrame or a structured numpy array.

```python

obj = {'xlabel': [3,4,5,6], 'ylabel': [2.4,13,123,11]}
plt.plot('xlabel', 'ylabel', data = obj)
```

#### 在一个图表上绘制多条数据
多次调用plot

```python
x1 = [1.2,3.2,6]
y1 = [10,11,23]
x2 = [133, 103]
y2 = [123, 111]
plt.plot(x1, y1, 'bo')
plt.plot(x2, y2, 'go')
```

```python
import numpy as np
x = [1, 2, 3]
y = np.array([[1, 2], [3, 4], [5, 6]])
plt.plot(x, y)

plt.plot(x, y)
```

```python
for col in range(y.shape[1]):
    plt.plot(x, y[:, col])
```

```python
plt.subplot(1, 2, 1)
ax = plt.gca()
ax.plot([1, 2, 3], [0, 0.5, 0.2])

plt.subplot(1, 2, 2)
ax = plt.gca()
ax.plot([3, 2, 1], [0, 0.5, 0.2])
```
